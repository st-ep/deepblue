```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
L = 100
sol = pd.read_csv('scaled_solution.csv')

sol.head()
```

```python
sol.describe().transpose()
```

```python
fuel_val_real = []
for k in range(0, L):
    samples = random.choices(sol['Fuel Value'].values, k = len(sol))
    fuel_val_real.append(np.average(samples))
    
plt.hist(fuel_val_real, alpha=0.8)
```
